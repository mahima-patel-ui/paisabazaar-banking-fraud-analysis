DATA CLEANING & PREPROCESSING

```python
df.isnull().sum()  # checking for null values


# Fill numric missing values with median
df.fillna(df.median(numeric_only=True),inplace=True)


```

```python
df.replace('NA',np.nan,inplace=True) # Handle missing values
df.fillna(df.median(numeric_only=True),inplace=True)
```

```python
df.duplicated().sum() # checking for duplicate values
```

```python
df.drop_duplicates(inplace=True) # droping duplicates
```

Droping Irrelevant columns

```python
#Droping coloumns that don,t add value to analysis
df.drop(columns=['ID','Customer_ID','Name','SSN'],inplace =True)
```

Clean And Standardize columns

```python
# Clean categorical columns
df['Credit_Score'] = df['Credit_Score'].str.strip().str.capitalize()
df['Payment_of_Min_Amount'] = df['Payment_of_Min_Amount'].str.strip().str.capitalize()
df['Credit_Mix'] = df['Credit_Mix'].str.strip().str.capitalize()
```

Converting data types

```python
# convert month to int and ensure numric fields are correct
df['Month'] = df['Month'].astype(int)
df['Credit_History_Age'] = pd.to_numeric(df['Credit_History_Age'],errors='coerce')
```

EXPLORATORY DATA ANALYSIS

1 . Credit Score distribution

```python
sns.countplot(data=df,x='Credit_Score')
plt.title("Distribution of Credit Score")
plt.show()
```

Insight:

The majority of individuals fall under the “standard” credit score category, followed by “poor”, while “good” credit scores form the smallest group.

This indicates a right-skewed credit score distribution, suggesting that most customers maintain unhealthy credit behavior.

However, a notable number of customers  have “Standard” or “Poor” scores, providing room for targeted improvement programs.

2. AGE VS Credit SCore

```python
sns.histplot(data=df,x='Age',hue='Credit_Score')
plt.title('credit score distribution by age')
plt.show
```

Insight:

Individuals in the 30–50 age range generally have higher (Good) credit scores.

Younger individuals (below 25) show a wider spread toward “Poor” and “Standard” scores — possibly due to limited credit history or recent credit activity.

Older individuals (50+) tend to have fewer “Poor” scores, reflecting more financial stability and experience.

3. Annual income vs Credit score

```python
sns.boxplot(data=df,x='Credit_Score',y='Annual_Income')
plt.title('Annual Income vs Credit Score')
plt.show()
```

Insight:

Customers with Good credit scores have significantly higher median annual incomes compared to those with “Standard” or “Poor” scores.

Higher income contributes to better repayment capability and lower credit utilization, improving overall credit health.

However, the presence of some “Poor” scorers even at higher incomes indicates that behavioral factors (like delayed payments) still play a major role beyond income.

4. Credit Utilization Ratio vs CRredit score

```python
sns.barplot(data=df, x='Credit_Score', y='Credit_Utilization_Ratio')
plt.title("Credit Utilization Ratio by Credit Score Category")
plt.show()
```

Insight:

Graph shows there is no much effect of credit utilization ratio.Good,standard and poor have same length bar showing their is no as such changes on  credit utization ratio

5. Delayed Payment vs Credit Score

```python
sns.barplot(data=df, x='Credit_Score', y='Num_of_Delayed_Payment')
plt.title("Number of Delayed Payments vs Credit Score")
plt.show()
```

Insight:

There is a strong negative correlation between delayed payments and credit scores.

“Poor” credit scorers have, on average, the highest number of delayed payments, while “Good” scorers have the fewest.

Timely payments are the most direct and powerful factor affecting an individual’s credit rating.

6 Number of Credit Cards vs Credit Score

```python
sns.boxplot(data=df, x='Credit_Score', y='Num_Credit_Card')
plt.title("Number of Credit Cards vs Credit Score")
plt.show()
```

Insight:

Individuals with Good credit scores tend to have multiple credit cards, but manage them responsibly (with low utilization).

“Poor” scorers often have fewer cards, possibly relying on single high-utilization accounts.

Having multiple cards is beneficial only when utilization and payments are well-managed.

7. Monthly Trend in Credit score

```python
monthly_trend = df.groupby("Month")["Credit_Score"].value_counts(normalize=True).unstack()
monthly_trend.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title("Monthly Trend of Credit Score Categories")
plt.ylabel("Proportion")
plt.show()
```

Insight:

Credit score distributions remain relatively stable across months, but some seasonal fluctuations may occur.

Slight dips in “Good” scores might align with festive or high-expense months, where short-term utilization spikes.

Consistency across months implies that credit score variation is more person-specific than seasonal.

8 Correlation Heatmap

```python
plt.figure(figsize=(12,6))
sns.heatmap(df.corr(numeric_only=True), cmap="coolwarm", annot=False)
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```

nsight:

Strong positive correlations:

Between Annual_Income and Monthly_Inhand_Salary.

Between Outstanding_Debt and Credit_Utilization_Ratio.

Strong negative correlations:

Between Num_of_Delayed_Payment and Credit_Score (encoded numerically).

Between Credit_Utilization_Ratio and Credit_Score.

Suggests that payment discipline and utilization efficiency are the two most important behavioral metrics.

#**🧾 Final Insights Summary — Paisabazaar Credit Score Analysis**


🔹 Key Findings


Customers with higher annual income generally maintain better credit scores.

Credit Utilization Ratio (CUR) is one of the strongest predictors —

Low CUR → Good credit score

High CUR → Poor credit score

Payment history has the most direct impact — frequent delayed payments drastically reduce scores.

Individuals aged 30–50 years tend to have healthier credit profiles compared to younger customers (<25).

Older individuals (50+) also show stable, moderate credit behavior with fewer payment issues.

Multiple credit cards, when used responsibly, correlate with higher credit scores.



Occupation-wise trend:

Professionals, Scientists, and Engineers → mostly “Good” scores

Entrepreneurs, Clerical workers, and Laborers → more “Standard” or “Poor” scores

Seasonal stability:

Credit score distribution remains mostly consistent across months

Slight dips in “Good” scores may occur in high-expense months (festive periods).

Correlation matrix highlights:

Strong positive link between income and in-hand salary

Strong negative link between delayed payments / utilization ratio and credit score

🔹 Behavioral & Financial Patterns

Income ≠ good credit by itself — repayment behavior and utilization play a bigger role.

Customers with moderate income but disciplined usage often achieve high scores.

Poor scorers usually display a combination of high utilization and frequent delays.

Credit diversity (multiple cards/loans managed responsibly) strengthens financial profiles.

🔹 Business Implications for Paisabazaar

Focus on behavior-based scoring models (utilization + payment patterns) rather than only income.

Introduce credit awareness programs to help customers maintain utilization below 30%.

Target younger users with financial literacy tools and smart repayment reminders.

Offer occupation-specific credit products (e.g., flexible loans for freelancers/laborers).

Reward on-time payers through loyalty programs or lower interest rates.

Use these insights to enhance customer segmentation and risk prediction accuracy.

#**🏁 Conclusion**

The analysis shows that payment discipline and credit utilization are the most critical factors influencing credit scores, outweighing income alone. Customers with timely payments and low credit usage maintain stronger credit profiles. These insights can help Paisabazaar enhance credit risk assessment, design personalized products, and promote responsible financial behavior among customers.
